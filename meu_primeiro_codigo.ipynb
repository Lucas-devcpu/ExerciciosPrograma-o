```
#visualização de dados em python

import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]
z = [200,25,4000,330,100]

titulo = "Scatterplot: Gráfico de dispersão"
eixox = "Eixo x"
eixoy = "Eixo y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label = "Meus Pontos", color = '#F08080', marker=".", s=z)
plt.plot(x,y, color= "k", linestyle="--", linewidth="5")
plt.legend()

plt.show()
```

```
#visualização de dados em python

import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]
z = [200,25,4000,330,100]

titulo = "Scatterplot: Gráfico de dispersão"
eixox = "Eixo x"
eixoy = "Eixo y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label = "Meus Pontos", color = '#F08080', marker=".", s=z)
plt.plot(x,y, color= "#995099", linestyle="--")
plt.legend()

plt.savefig("figura1.png", dpi=300)
#plt.show()
```
